## scraping + imports

```python
#imports
import pandas as pd
import numpy as np
from datetime import datetime
from tqdm import tqdm_notebook, tqdm
import urllib.request
import os
#from to dates
res = pd.date_range(start="2022-01-01",end="2022-11-27")

# #looping over all days individually

for date in tqdm(res):
    day = date.strftime('%Y-%m-%d') 
    url = f"https://datahub.ren.pt/service/download/csv/1363?startDateString={day}&endDateString={day}&culture=en-GB"
    result = urllib.request.urlretrieve(url, f"./Raw_data/{day}.csv") #saving into the Raw_data folder
    
```
